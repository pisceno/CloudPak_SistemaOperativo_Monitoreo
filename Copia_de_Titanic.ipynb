```python
import pandas as pd
url = "https://raw.githubusercontent.com/juarezefren/datasets/main/titanic_train.csv"
df = pd.read_csv(url)
df
```

```python
df.info()
```

```python
df['Survived'].value_counts()
```

```python
df['Age'].unique()
```

```python
valor = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(valor)
```

```python
df['Cabin'] = df['Cabin'].fillna('desconocido')
```

```python
valor = df['Age'].mean()
df['Age'] = df['Age'].fillna(valor)
```

```python
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()

df['Name'] = encoder.fit_transform(df[['Name']])
df['Sex'] = encoder.fit_transform(df[['Sex']])
df['Ticket'] = encoder.fit_transform(df[['Ticket']])
df['Cabin'] = encoder.fit_transform(df[['Cabin']])
df['Embarked'] = encoder.fit_transform(df[['Embarked']])

df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df['Pclass'] = scaler.fit_transform(df[['Pclass']])
df['Name'] = scaler.fit_transform(df[['Name']])
df['Sex'] = scaler.fit_transform(df[['Sex']])
df['Age'] = scaler.fit_transform(df[['Age']])
df['SibSp'] = scaler.fit_transform(df[['SibSp']])
df['Parch'] = scaler.fit_transform(df[['Parch']])
df['Ticket'] = scaler.fit_transform(df[['Ticket']])
df['Cabin'] = scaler.fit_transform(df[['Cabin']])
df['Embarked'] = scaler.fit_transform(df[['Embarked']])
df['Fare'] = scaler.fit_transform(df[['Fare']])

df.head()
```

```python
y = df['Survived']
print(y.head())
X = df.drop(columns=['Survived','PassengerId'])
print(X.head())
```

```python
X.info()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE

modelo = LogisticRegression(solver='lbfgs', multi_class='auto', max_iter=1000)

rfe = RFE(estimator=modelo, n_features_to_select=6, step=1)

rfe = rfe.fit(X, y)

print(rfe.support_)
print(rfe.ranking_)
X_selected = rfe.transform(X)
print(X_selected)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Entrena un clasificador de bosque aleatorio
clf = RandomForestClassifier()
clf.fit(X, y)

# Obtiene la importancia de las caracterÃ­sticas
importances = clf.feature_importances_

# Crea un DataFrame para visualizar las importancias
feature_importances = pd.DataFrame({'feature': X.columns, 'importance': importances})

# Ordena las caracterÃ­sticas por importancia
sorted_features = feature_importances.sort_values(by='importance', ascending=False)

print(sorted_features)
```

```python
# Calcular la correlaciÃ³n de cada caracterÃ­stica con la variable objetivo
correlations = df.corr()['Survived'].drop('Survived')

# Ordenar las caracterÃ­sticas por su correlaciÃ³n con la variable objetivo
sorted_features = correlations.abs().sort_values(ascending=False)

print(sorted_features)
```

```python
from sklearn.neighbors import KNeighborsClassifier
modelo = KNeighborsClassifier(n_neighbors=5)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(modelo, X_selected, y, cv=5)
print(scores)
print(scores.mean())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=21)
modelo.fit(X_train, y_train)

y_predicted = modelo.predict(X_test)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_predicted )
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score
from sklearn.metrics import f1_score

print("F1: ", f1_score(y_test, y_predicted))
print("Accuracy: ", accuracy_score(y_test, y_predicted))
print("Precision: ", precision_score(y_test, y_predicted))
print("Recall: ",recall_score(y_test, y_predicted))
```
