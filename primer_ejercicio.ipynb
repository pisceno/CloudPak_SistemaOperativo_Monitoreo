```python
import pandas as pd
url = "https://raw.githubusercontent.com/juarezefren/datasets/refs/heads/main/insects-15.csv"
df = pd.read_csv(url)
df.head()
```

```python
X=df[['abdomen', 'antena']]
X.head(5)
```

```python
y=df[['clase']]
y.head(5)
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(X,y)
```

```python
nuevo_ejemplo=[[5.1, 7.0]]
prediccion=model.predict(nuevo_ejemplo)
print(prediccion)
```

```python
X_test=df.iloc[10:,[0,2]]
X_test
```

```python
prediccion=model.predict(X_test)
print(prediccion)
```
